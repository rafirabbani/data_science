```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

<h2>No 1 : Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig,ax = plt.subplots(ncols=3, sharey=True, figsize=[15,8])
ax[0].plot(x, tan, color= 'b', label='tanx')
ax[1].plot(x, cos, color = 'r', label='cosx')
ax[2].plot(x, sin, color= 'g', label='sinx')
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik cosx')
ax[2].set_title('Grafik sinx')
fig.legend(loc= 'upper center')
plt.show()
```

<h2>No 2 : Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=[15,7])
ax.plot(x, y, 'b')
ax.set_title('Grafik Cos X')

ins1 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins2 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins3 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins4 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])

ins1.plot(x, y2, 'y', x, y)
ins2.plot(x, y3, 'g', x, y)
ins3.plot(x, y4, 'k', x, y)
ins4.plot(x, y5, 'r', x, y)

ins1.set_title('Grafik CosX^2')
ins2.set_title('Grafik CosX^3')
ins3.set_title('Grafik CosX^4')
ins4.set_title('Grafik CosX^5')

plt.show()
```
