```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Quiz 1 Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
genre_group = df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales', 'Genre']].groupby(by='Genre').mean()
genre_trans = genre_group.transpose()
genre_trans
```

<h2>Perbandingan Kuantitatif Barplot : Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
x = np.arange(len(genre_trans.index))
labels = genre_trans.columns.values
z = 0
data_list = []

while z < len(labels) :
    data_list.append(genre_trans[labels[z]])
    z += 1
    
width = 0.05
j = 0


fig, ax = plt.subplots(figsize=[16,8])

while j < len(data_list) :
    ax.bar(x + j * width, data_list[j], width, label=labels[j])
    j += 1

ax.set_xticks(x + 5.5 * width)
ax.set_xticklabels(genre_trans.index)
ax.set_title('Mean Sales Video Game by Genre')
ax.set_ylabel('Mean Sales')
ax.set_xlabel('Region Sales')

fig.legend()

plt.show()

print(data_list)
```
