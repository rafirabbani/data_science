```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('vgsales.csv')
df.head()
```

```python
year_data = df['Year'] > 2000
sell_2000 = df[(year_data)]
genre_data = sell_2000['Genre'].value_counts()
cat = genre_data.index
freq = genre_data.values

fig, ax = plt.subplots(figsize=[15,10])
ax.bar(cat, freq)
ax.set_title('Penjualan Videogame Berdasarkan Genre dari Tahun 2000 - 2017')
ax.set_ylabel('Jumlah Penjualan')
ax.set_xlabel('Genre Videogame')
plt.show()
```

```python
ps4 = sell_2000['Platform'] == 'PS4'
ps4_data = sell_2000[(ps4)]

group_year = ps4_data.groupby(by='Year').sum().reset_index()

fig, ax = plt.subplots(figsize=[10,5])
ax.plot(group_year['Year'], group_year['Global_Sales'], marker='o')
ax.set_title('Penjualan Game pada Platform PS4 2013 - 2017')
ax.set_ylabel('Penjualan')
ax.set_xlabel('Tahun')
plt.show()
```
