```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)

figure = plt.figure(figsize=[15,7])
axes = figure.add_subplot()

def plot_func(x,y) :
    return axes.plot(x,y)

plot_func(x,y)
plot_func(x,y2)
plot_func(x,y3)

plt.show()
```
